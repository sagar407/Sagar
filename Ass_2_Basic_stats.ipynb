```python
company_name = ['Allied Signal','Bankers Trust','General Mills','ITT Industries','J.P.Morgan & Co.','Lehman Brothers',
      'Marriott','MCI','Merrill Lynch','Microsoft','Morgan Stanley','Sun Microsystems','Travelers','US Airways',
      'Warner-Lambert']
measure_x = pd.Series([24.23,25.53,25.41,24.14,29.62,28.25,25.81,24.39,40.26,32.95,91.36,25.99,39.42,26.71,35.00])
```

```python
plt.figure(figsize = (10,10))
plt.pie(measure_x, labels = company_name,shadow = False,
     textprops = {'size':'medium','color':'black'})
plt.title('Name of Companies with respect to measure X', fontweight = 'bold',fontsize = 22, color='Blue')
#plt.savefig('Question1piechart.png')
plt.show()
```

```python
#boxplot
sns.boxplot(x=measure_x)
plt.show()
```

```python
measure_x=pd.Series(measure_x)
```

```python
print('First Quantile Q1  = ',np.quantile(measure_x,0.25))
print('second Quantile Q2 = ',np.quantile(measure_x,0.50))
print('third Quantile Q3  = ',np.quantile(measure_x,0.75))
print('median             = ',np.median(measure_x))
```

```python
print('mean of measure_x      is  μ    :',round(measure_x.mean(),4))
print('std of measure_x       is  σ    :',round(measure_x.std(),4))
print('varience of measure_x  is  σ^2  :',round(measure_x.var(),4))
```
